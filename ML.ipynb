Treating data imbalance

```python
test_dataframe = dataframe

value_counts_series = test_dataframe["COLLEGE"].value_counts()
df_value_counts = value_counts_series.reset_index()
df_value_counts.columns = ['COLLEGE', 'COUNT']
```

```python
college_included = 0
college_not_included = 0

# Loop through the DataFrame
for index, row in df_value_counts.iterrows():
    college_name = row['COLLEGE']
    count_value = row['COUNT']

    print(f"College: {college_name}, Data: {count_value}")

    if count_value > target_threshould:
      college_included = college_included + 1

      num_rows_to_drop = count_value - target_threshould
      print("To randomly remove: " + str(num_rows_to_drop))

      array_rand = test_dataframe[test_dataframe['COLLEGE'] == college_name].index.values

      for _ in range(num_rows_to_drop):
        array_index = random.choice(range(len(array_rand)))
        actual_index = array_rand[array_index]

        test_dataframe = test_dataframe.drop(actual_index)

        array_rand = test_dataframe[test_dataframe['COLLEGE'] == college_name].index.values

    else:
      test_dataframe = test_dataframe[test_dataframe['COLLEGE'] != college_name]
      college_not_included = college_not_included + 1
      print("NOT INCLUDED -- " + str(count_value))

    print("----")


print("\n")

print("Original dataset   : " + str(dataframe.shape))
print("New dataset        : " + str(test_dataframe.shape))

print("\n")

print(str(college_included) + " colleges included")
print(str(college_not_included) + " colleges not included")

test_dataframe = test_dataframe.reset_index(drop=True)
dataframe = test_dataframe
```
